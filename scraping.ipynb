```python
import requests
from bs4 import BeautifulSoup
import re
keyYlist = ['a', 'b']
for k in keyYlist:
    for l in range(1, 100):
        URL = f"https://www.merriam-webster.com/browse/medical/{k}/{l}"
        page = requests.get(URL)
        if page.status_code == 200:
            print(URL)
            soup = BeautifulSoup(page.content, "html.parser")
            myele = soup.find_all("a", attrs={"class": "pb-4 pr-4 d-block"})
            f = open("text.txt", "a")
            for i in range(len(myele)):
                a = myele[i].text
                f.write(a)
                f.write("\n")
```
